```python
import torch
from torch import nn
```

## Initial Models

```python
class Net_A(nn.Module):
  def __init__(self):
    super().__init__()

    self.conv_3to64 = nn.Conv2d(3, 64, 3)
    self.conv__64to128 = nn.Conv2d(64, 128, 3)
    self.conv_128to256 = nn.Conv2d(128, 256, 3)
    self.conv_256to256 = nn.Conv2d(256, 256, 3)
    self.conv_256to512 = nn.Conv2d(256, 512, 3)
    self.conv_512to512 = nn.Conv2d(512, 512, 3)
    self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
    self.fc1 = nn.Linear(512*3*3, 4096)
    self.fc2 = nn.Linear(4096, 4096)
    self.fc3 = nn.Linear(4096, 1000)
    self.logsoftmax = nn.LogSoftmax(dim = 1)

  def forward(self, x):
    x = self.conv_3to64(x)
    x = self.pool(x)
    x = self.conv__64to128(x)
    x = self.pool(x)
    x = self.conv_128to256(x)
    x = self.conv_256to256(x)
    x = self.pool(x)
    x = self.conv_256to512(x)
    x = self.conv_512to512(x)
    x = self.pool(x)
    x = self.conv_512to512(x)
    x = self.conv_512to512(x)
    x = self.pool(x)
    x = x.view(1, -1)
    x = self.fc1(x)
    x = self.fc2(x)
    x = self.fc3(x)
    x = self.logsoftmax(x)

    return x 
```

```python
class Net_B(nn.Module):
  def __init__(self):
    super().__init__()

    self.conv_3to64 = nn.Conv2d(3, 64, 3)
    self.conv_64to64 = nn.Conv2d(64, 64, 3)
    self.conv_64to128 = nn.Conv2d(64, 128, 3)
    self.conv_128to128 = nn.Conv2d(128, 128, 3)
    self.conv_128to256 = nn.Conv2d(128, 256, 3)
    self.conv_256to256 = nn.Conv2d(256, 256, 3)
    self.conv_256to512 = nn.Conv2d(256, 512, 3)
    self.conv_512to512 = nn.Conv2d(512, 512, 3)
    self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
    self.fc1 = nn.Linear(512*3*3, 4096)
    self.fc2 = nn.Linear(4096, 4096)
    self.fc3 = nn.Linear(4096, 1000)
    self.logsoftmax = nn.LogSoftmax(dim = 1)

  def forward(self, x):
    x = self.conv_3to64(x)
    x = self.conv_64to64(x)
    x = self.pool(x)
    x = self.conv_64to128(x)
    x = self.conv_128to128(x)
    x = self.pool(x)
    x = self.conv_128to256(x)
    x = self.conv_256to256(x)
    x = self.pool(x)
    x = self.conv_256to512(x)
    x = self.conv_512to512(x)
    x = self.pool(x)
    x = self.conv_512to512(x)
    x = self.conv_512to512(x)
    x = self.pool(x)
    x = x.view(1, -1)
    x = self.fc1(x)
    x = self.fc2(x)
    x = self.fc3(x)
    x = self.logsoftmax(x)

    return x 
```

```python
class Net_C(nn.Module):
  def __init__(self):
    super().__init__()

    self.conv_3to64 = nn.Conv2d(3, 64, 3)
    self.conv_64to64 = nn.Conv2d(64, 64, 3)
    self.conv_64to128 = nn.Conv2d(64, 128, 3)
    self.conv_128to128 = nn.Conv2d(128, 128, 3)
    self.conv_128to256 = nn.Conv2d(128, 256, 3)
    self.conv_256to256 = nn.Conv2d(256, 256, 3)
    self.conv_256to512 = nn.Conv2d(256, 512, 3)
    self.conv_512to512 = nn.Conv2d(512, 512, 3)
    self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
    self.fc1 = nn.Linear(512*1*1, 4096)
    self.fc2 = nn.Linear(4096, 4096)
    self.fc3 = nn.Linear(4096, 1000)
    self.logsoftmax = nn.LogSoftmax(dim = 1)

  def forward(self, x):
    x = self.conv_3to64(x)
    x = self.conv_64to64(x)
    x = self.pool(x)
    x = self.conv_64to128(x)
    x = self.conv_128to128(x)
    x = self.pool(x)
    x = self.conv_128to256(x)
    x = self.conv_256to256(x)
    x = self.conv_256to256(x)
    x = self.pool(x)
    x = self.conv_256to512(x)
    x = self.conv_512to512(x)
    x = self.conv_512to512(x)
    x = self.pool(x)
    x = self.conv_512to512(x)
    x = self.conv_512to512(x)
    x = self.conv_512to512(x)
    x = self.pool(x)
    x = x.view(1, -1)
    x = self.fc1(x)
    x = self.fc2(x)
    x = self.fc3(x)
    x = self.logsoftmax(x)

    return x 
```

```python
class Net_D(nn.Module):
  def __init__(self):
    super().__init__()

    self.conv_3to64 = nn.Conv2d(3, 64, 3)
    self.conv_64to64 = nn.Conv2d(64, 64, 3)
    self.conv_64to128 = nn.Conv2d(64, 128, 3)
    self.conv_128to128 = nn.Conv2d(128, 128, 3)
    self.conv_128to256 = nn.Conv2d(128, 256, 3)
    self.conv_256to256 = nn.Conv2d(256, 256, 3)
    self.conv_256to512 = nn.Conv2d(256, 512, 3)
    self.conv_512to512 = nn.Conv2d(512, 512, 3)
    self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
    self.fc1 = nn.Linear(512*1*1, 4096)
    self.fc2 = nn.Linear(4096, 4096)
    self.fc3 = nn.Linear(4096, 1000)
    self.logsoftmax = nn.LogSoftmax(dim = 1)

  def forward(self, x):
    x = self.conv_3to64(x)
    x = self.conv_64to64(x)
    x = self.pool(x)
    x = self.conv_64to128(x)
    x = self.conv_128to128(x)
    x = self.pool(x)
    x = self.conv_128to256(x)
    x = self.conv_256to256(x)
    x = self.conv_256to256(x)
    x = self.pool(x)
    x = self.conv_256to512(x)
    x = self.conv_512to512(x)
    x = self.conv_512to512(x)
    x = self.pool(x)
    x = self.conv_512to512(x)
    x = self.conv_512to512(x)
    x = self.conv_512to512(x)
    x = self.pool(x)
    x = x.view(1, -1)
    x = self.fc1(x)
    x = self.fc2(x)
    x = self.fc3(x)
    x = self.logsoftmax(x)

    return x 
```

## Execute Models

```python
model_result_dic = {}

model_dic = {"A":Net_A(), "B":Net_B(), "C":Net_C() , "D":Net_D()}

for key , model in model_dic.items():
    print(f"Model {key} => {model} ")
    random_sample = torch.randn((3, 224, 224))
    x = model(random_sample)
    model_result_dic[key] = x
```
