```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python

# parameters
num_samples = 1000
time = np.linspace(0, 10, num_samples)
frequency = 5  
amplitude = 1  
noise_amplitude = 0.5 

# Create Sin Data
sinusoidal_data = amplitude * np.sin(2 * np.pi * frequency * time)

# add noise to data
noisy_data = sinusoidal_data + noise_amplitude * np.random.randn(num_samples)

# denoising
window_size = 11
filtered_data = signal.medfilt(noisy_data, kernel_size=window_size)

```

```python

plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.plot(time, sinusoidal_data, label='Sinusoidal Data')
plt.title('Original Sinusoidal Data')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(time, noisy_data, color='r', label='Noisy Data')
plt.title('Noisy Sinusoidal Data')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(time, filtered_data, color='g', label='Filtered Data')
plt.title('Filtered Sinusoidal Data')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()
```
