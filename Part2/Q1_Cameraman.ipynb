```python
import matplotlib.pyplot as plt
from skimage import data
from skimage.filters import gaussian , unsharp_mask
```

## Orginal --> Blur --> Unsharp Image (Convolution)

```python
camera_image = data.camera() #orginal image

# conv blur image
kernel_size = 5
sigma = 1
blurred_image = camera_image
for _ in range(5):
    blurred_image = gaussian(blurred_image, sigma=sigma)

# unsharp image for reconstruct the orginal image
unsharped_image = blurred_image
for _ in range(10):
    unsharped_image = unsharp_mask(unsharped_image )
```

```python

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(camera_image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(blurred_image, cmap='gray')
plt.title('Blurred Image')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(unsharped_image, cmap='gray')
plt.title('UnSharp Image (Aboute Orginal Image)')
plt.axis('off')
```
